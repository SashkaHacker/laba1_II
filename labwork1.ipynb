```python
import numpy as np
data_path = "seeds_dataset.txt"
data = np.genfromtxt(data_path, delimiter="\t")
print(data)
```

### Тип переменной и форма (shape)

```python
print ("Data type : ", type(data))
print ("Data shape : ", data.shape)
print (data[:10])
```

### Построение графиков с использованием Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

# Данные из отдельных столбцов
area = [] # площадь семени
perimeter = [] # периметр семени
compactness = [] # компактность семени
length_of_kernel = [] # длина семени
width_of_kernel = [] # ширина семени
asymmetry_coefficient = [] # коэффициент асимметрии семени
length_of_kernel_groove = [] # длина борозды семени



# Выполняем обход всей коллекции data2
for dot in data:
    area.append(dot[0])
    perimeter.append(dot[1])
    compactness.append(dot[2])
    length_of_kernel.append(dot[3])
    width_of_kernel.append(dot[4])
    asymmetry_coefficient.append(dot[5])
    length_of_kernel_groove.append(dot[6])

print(f'Поле area: \nМинимальное значение: {min(area)}\nМаксимальное значение: {max(area)}\nСреднее значение: {sum(area) / len(area)}\n')
print(f'Поле perimeter: \nМинимальное значение: {min(perimeter)}\nМаксимальное значение: {max(perimeter)}\nСреднее значение: {sum(perimeter) / len(perimeter)}\n')
print(f'Поле compactness: \nМинимальное значение: {min(compactness)}\nМаксимальное значение: {max(compactness)}\nСреднее значение: {sum(compactness) / len(compactness)}\n')
print(f'Поле length_of_kernel: \nМинимальное значение: {min(length_of_kernel)}\nМаксимальное значение: {max(length_of_kernel)}\nСреднее значение: {sum(length_of_kernel) / len(length_of_kernel)}\n')
print(f'Поле compactness: \nМинимальное значение: {min(width_of_kernel)}\nМаксимальное значение: {max(width_of_kernel)}\nСреднее значение: {sum(width_of_kernel) / len(width_of_kernel)}\n')
print(f'Поле asymmetry_coefficient: \nМинимальное значение: {min(asymmetry_coefficient)}\nМаксимальное значение: {max(asymmetry_coefficient)}\nСреднее значение: {sum(asymmetry_coefficient) / len(asymmetry_coefficient)}\n')
print(f'Поле length_of_kernel_groove: \nМинимальное значение: {min(length_of_kernel_groove)}\nМаксимальное значение: {max(length_of_kernel_groove)}\nСреднее значение: {sum(length_of_kernel_groove) / len(length_of_kernel_groove)}\n')




```

```python
# Строим графики по проекциям данных
# Учитываем, что каждые 70 типов зерен идут последовательно
plt.figure(1)
kama, = plt.plot(length_of_kernel[:70], width_of_kernel[:70], 'ro', label='Kama')
rosa, = plt.plot(length_of_kernel[70:140], width_of_kernel[70:140], 'g^', label='Rosa')
canadian, = plt.plot(length_of_kernel[140:210], width_of_kernel[140:210], 'bs', label='Canadian')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Длина семени')
plt.ylabel('Ширина семени')


plt.figure(2)
kama, = plt.plot(asymmetry_coefficient[:70], length_of_kernel_groove[:70], 'ro', label='Kama')
rosa, = plt.plot(asymmetry_coefficient[70:140], length_of_kernel_groove[70:140], 'g^', label='Rosa')
canadian, = plt.plot(asymmetry_coefficient[140:210], length_of_kernel_groove[140:210], 'bs', label='Canadian')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Коэффициент асимметрии семени')
plt.ylabel('Длина борозды семени')


plt.figure(3)
kama, = plt.plot(compactness[:70], perimeter[:70], 'ro', label='Kama')
rosa, = plt.plot(compactness[70:140], perimeter[70:140], 'g^', label='Rosa')
canadian, = plt.plot(compactness[140:210], perimeter[140:210], 'bs', label='Canadian')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Компактность семени')
plt.ylabel('Периметр семени')

plt.show()
```
